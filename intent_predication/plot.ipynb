```python
import numpy as np
import matplotlib.pyplot as plt

models = ['DeBERTa-base', 'DistilBERT-base']

# Original latencies
avg_batch_latency_ms = [163.54, 112.24]  # ms
total_latency_ms = [25676.41, 17621.36]  # ms
avg_latency_per_item_ms = [2.57, 1.76]  # ms

# Convert Avg Batch Latency and Total Latency to seconds
avg_batch_latency_sec = [x / 1000 for x in avg_batch_latency_ms]
total_latency_sec = [x / 1000 for x in total_latency_ms]

x = np.arange(len(models))  # label locations
width = 0.25  # width of bars

fig, ax = plt.subplots(figsize=(10,6))

bar1 = ax.bar(x - width, avg_batch_latency_sec, width, label='Avg Batch Latency (seconds)')
bar2 = ax.bar(x, total_latency_sec, width, label='Total Latency (seconds)')
bar3 = ax.bar(x + width, avg_latency_per_item_ms, width, label='Avg Latency per Item (ms)')

# Add labels and title
ax.set_ylabel('Latency')
ax.set_title('Latency Comparison: DeBERTa-base vs DistilBERT-base')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()

# Add values on top of bars
def add_labels(bars, is_seconds=False):
    for bar in bars:
        height = bar.get_height()
        label = f'{height:.2f}s' if is_seconds else f'{height:.2f}ms'
        ax.annotate(label,
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0,3),
                    textcoords="offset points",
                    ha='center', va='bottom')

add_labels(bar1, is_seconds=True)
add_labels(bar2, is_seconds=True)
add_labels(bar3, is_seconds=False)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

models = ['DeBERTa-base', 'DistilBERT-base']

# Latency data
avg_batch_latency_ms = [163.54, 112.24]     # ms
total_latency_ms = [25676.41, 17621.36]     # ms
avg_latency_per_item_ms = [2.57, 1.76]      # ms

# Convert batch and total latency to seconds
avg_batch_latency_s = [x / 1000 for x in avg_batch_latency_ms]
total_latency_s = [x / 1000 for x in total_latency_ms]

x = np.arange(len(models))  # label locations
width = 0.35  # width of bars

fig, ax1 = plt.subplots(figsize=(10,6))

# Plot avg batch latency and total latency on primary y-axis (seconds)
bar1 = ax1.bar(x - width/2, avg_batch_latency_s, width, label='Avg Batch Latency (s)', color='skyblue')
bar2 = ax1.bar(x + width/2, total_latency_s, width, label='Total Latency (s)', color='steelblue')
ax1.set_ylabel('Latency (seconds)')
ax1.set_xticks(x)
ax1.set_xticklabels(models)
ax1.set_title('Latency Comparison: DeBERTa-base vs DistilBERT-base')
ax1.legend(loc='upper left')

# Secondary y-axis for average latency per item (ms)
ax2 = ax1.twinx()
bar3 = ax2.bar(x + width*1.5, avg_latency_per_item_ms, width*0.6, label='Avg Latency per Item (ms)', color='orange')
ax2.set_ylabel('Latency (milliseconds)')
ax2.legend(loc='upper right')

# Function to add labels on bars
def add_labels(bars, axis, is_seconds=False):
    for bar in bars:
        height = bar.get_height()
        label = f'{height:.2f}s' if is_seconds else f'{height:.2f}ms'
        axis.annotate(label,
                      xy=(bar.get_x() + bar.get_width() / 2, height),
                      xytext=(0, 3),
                      textcoords="offset points",
                      ha='center', va='bottom')

add_labels(bar1, ax1, is_seconds=True)
add_labels(bar2, ax1, is_seconds=True)
add_labels(bar3, ax2, is_seconds=False)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

models = ['DeBERTa-base', 'DistilBERT-base']

# Latency values
total_latency_s = [25676.41 / 1000, 17621.36 / 1000]  # seconds
avg_batch_latency_ms = [163.54, 112.24]  # milliseconds
avg_latency_per_item_ms = [2.57, 1.76]  # milliseconds

x = np.arange(len(models))
width = 0.35

# Plot 1: Total Latency in seconds
fig1, ax1 = plt.subplots(figsize=(8,5))
bars1 = ax1.bar(x, total_latency_s, width, color=['skyblue', 'orange'])

ax1.set_title('Total Latency (seconds)')
ax1.set_ylabel('Latency (s)')
ax1.set_xticks(x)
ax1.set_xticklabels(models)

for bar in bars1:
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, yval + 0.1, f'{yval:.2f}s', ha='center', va='bottom')

# Plot 2: Avg Batch Latency and Avg Latency per Item (milliseconds)
fig2, ax2 = plt.subplots(figsize=(8,5))

bars2_1 = ax2.bar(x - width/2, avg_batch_latency_ms, width, label='Avg Batch Latency (ms)', color='green')
bars2_2 = ax2.bar(x + width/2, avg_latency_per_item_ms, width, label='Avg Latency per Item (ms)', color='purple')

ax2.set_title('Average Latencies (milliseconds)')
ax2.set_ylabel('Latency (ms)')
ax2.set_xticks(x)
ax2.set_xticklabels(models)
ax2.legend()

for bar in bars2_1:
    yval = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{yval:.2f}ms', ha='center', va='bottom')

for bar in bars2_2:
    yval = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2, yval + 0.1, f'{yval:.2f}ms', ha='center', va='bottom')

plt.show()
```
